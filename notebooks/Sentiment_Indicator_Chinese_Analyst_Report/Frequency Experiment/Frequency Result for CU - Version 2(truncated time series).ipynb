# Objective
The purpose of this notebook is to study the relationship between standard deviation and number of recommendations in a day. This version we will truncate the time series so we can have more balance distribution of number of recommednations

# Experiment relationship between Frequency and Forward Volatility

```python
# Input: Series is the freq series of recommendation per day
# Obj: This fucntion will convert days with zero recommendation into int zero instead of nan 
def convert_zero_nan(series):
    result = []
    first = 1
    for i in series:
        if i != np.nan and first == 1:
            first = 0
        if i == np.nan and first ==0:
            result.append(0)
        else:
            result.append(i)
        
    return pd.Series(result)
```

```python
# Input: data is the total dataframe, target is the name of the target column and indicator is the name of the predictive column
# Obj: This fucntion will give a scatter plot to let us visualize the relationship
def experiment_result(target,indicator,data):
    X = data[indicator].values
    X = np.reshape(X, (-1,1))
    data.plot.scatter(y =target,x=indicator,title = target +" vs "+indicator)
    print("Correlation for " + target + ' is : ' + str(data[indicator].corr(data[target])))
    target = data[target]
    model = linear_model.LinearRegression().fit(X,target)
    predicted = model.predict(X)
    plt.plot(data[indicator],predicted, 'r', label='fitted line')
    return
```

```python
# Input: data is the total dataframe, vol_col is the name of the target volatility col
# Obj: This fucntion will give us the best splitting point by calculating the point biserial correlation between freq and vol as well as verifying the result 
# with annova table across all possible freq threshold  
def caluclate_best_threshold(data,vol_col):
    best = 0
    best_thres = 0
    freq = data['Freq']
    target = data[vol_col]  
    for i in freq.unique():
        
        #Calculate Point Bieserial Correlation
        test_data = freq.apply(lambda x: 0 if x<= i else 1)
        corr = stats.pointbiserialr(test_data, target)
        
        # Calculate annova
        test_data = pd.concat([test_data, target], axis=1)
        results = ols( vol_col + ' ~ C(Freq)', data=test_data).fit()
        accept = results.f_pvalue<0.05
        
        print('When threshold is ' + str(i) +" :",corr,accept)
        if np.abs(corr[0])> np.abs(best) and accept:
            best = corr[0]
            best_thres = i
    return (best,best_thres)
```

# Deviation from long term volatility

## Data Visualization for deviation from long term volatility

```python
vol = price[[u'Date', u'LMCADY', u'return_1d']].copy()
vol['Std_st'] = (vol[u'return_1d'].rolling(10).std())
vol['Std_lt'] = (vol[u'return_1d'].rolling(250).std())
vol['Std'] = (vol['Std_st']-vol['Std_lt'])
vol['std_1d'] = vol['Std'].shift(-1)
vol['std_3d'] = vol['Std'].shift(-3)
vol['std_5d'] = vol['Std'].shift(-5)
vol['std_7d'] = vol['Std'].shift(-7)
vol['std_10d'] = vol['Std'].shift(-10)
vol.dropna(inplace=True)
```

```python
sns.distplot(vol['Std'], hist=False, color="g", kde_kws={"shade": True}).set_title('Distribution')
```

```python
xmed = vol['Std'].mean()
ax = sns.lineplot(x="Date", y="Std", data=vol).set_title('Time Seires')
x = plt.gca().axes.get_xlim()
plt.plot(x,  len(x)*[xmed], sns.xkcd_rgb["pale red"])
sm.graphics.tsa.plot_acf(vol['Std'], lags=40)
plt.show()
```

## Correaltion between frequency and deviation from long term vol

```python
merged = pd.merge(vol, new_freq, left_on='Date', right_index=True,how = 'left')
merged['Freq'] = convert_zero_nan(merged['Freq'])
merged.dropna(inplace = True)
```

```python
experiment_result('std_1d','Freq',merged)
experiment_result('std_3d','Freq',merged)
experiment_result('std_5d','Freq',merged)
experiment_result('std_7d','Freq',merged)
experiment_result('std_10d','Freq',merged)
```

## Calculate the best split and annova table 

```python
caluclate_best_threshold(merged,'std_1d')
```

# Conclusion
1. There is a positive correlation between number of recommendation per day and deviation from long term volatility.
2. Using 7 as our threshold, we can have biggest correlation of 0.389
